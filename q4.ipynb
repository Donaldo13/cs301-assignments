```python
# Question 4a
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


#read from file 
df = pd.read_csv('nyc-east-river-bicycle-counts.csv')
# take the brooklyn data set and assign it to bridge array
bridge = df['Brooklyn Bridge'].tolist()


#plot bridge array into the histogram 
plt.hist(bridge , "auto", orientation = 'vertical' ,  ec='green') 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
avg = np.average(bridge)
s = np.random.poisson(avg, 1000)

count, bins, ignored = plt.hist(s, 14, density=True , ec ='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the arrival rates
lambdas = np.array([1, 3, 4])

# Set the service rate
mu = 4

# Set the maximum time for simulation
max_time = 100

# Set the time step for simulation
time_step = 0.1

# Initialize the list to store the number of requests in the queue
queue_lengths = []

# Set the initial number of requests in the queue
queue_length = 0

for t in np.arange(0, max_time, time_step):
    # Calculate the number of requests that arrive in the current time step
    arrivals = np.random.poisson(lambdas * time_step)
    queue_length += arrivals

    # Calculate the number of requests that are served in the current time step
    served_requests = np.random.exponential(mu * time_step)
    queue_length = queue_length - served_requests

    # Store the number of requests in the queue
    queue_lengths.append(queue_length)

# Plot the number of requests in the queue as a function of time
plt.plot(np.arange(0, max_time, time_step), queue_lengths)
plt.xlabel('Time (s)')
plt.ylabel('Number of requests in the queue')
plt.show()
```
